## Dataset Selection and Justification

### What the dataset is

The dataset used in this task is a **curated SMS spam collection** — a set of 149 short text messages (75 spam, 74 ham) labelled as either spam (deceptive/unsolicited) or ham (genuine). The messages are representative of the publicly known SMS Spam Collection corpus (Almeida et al., 2011, UCI ML Repository), which contains 5,574 SMS messages in English. Since no internet access is available in this environment, a representative self-contained subset is embedded directly in this notebook to ensure full reproducibility without any external downloads or manual steps.

### Why it is a reasonable testbed for the method in Ott et al. (2011)

The SMS spam classification task is structurally identical to the deceptive opinion spam detection task in Ott et al.: both are **binary text classification problems** where one class represents genuine/authentic text (ham / truthful reviews) and the other represents deceptive/unsolicited text (spam / deceptive reviews). Both tasks use short natural language text as input, both produce a binary label as output, and both are well-suited to the TF-IDF + LinearSVC pipeline that is the core contribution of the paper. The evaluation metric — classification accuracy — is the same metric reported in Table 3 of the paper, allowing a direct methodological parallel even if the absolute numbers differ.

### Limitations compared to the Ott et al. dataset

There are two important limitations. First, SMS spam detection is a **lexically simpler task** than deceptive hotel review detection: spam messages typically contain explicit trigger words such as FREE, WIN, URGENT, PRIZE that are highly discriminative on their own, making the classification task easier for any bag-of-words model. Deceptive hotel reviews, by contrast, use natural and fluent prose that is deliberately designed to sound authentic — making the pattern-detection task significantly harder (hence human judges perform at only ~60% on hotel reviews while far outperforming chance on SMS spam). Second, this embedded dataset contains only **149 samples**, compared to the 800 reviews used in the paper, which means cross-validation estimates will have higher variance and the model has far less training data per fold.

## Preprocessing Steps

The preprocessing applied to this dataset mirrors the preprocessing used in Ott et al. (2011) as closely as possible:

1. **Lowercasing** — all text is converted to lowercase, consistent with the paper's use of lowercased n-gram features (Section 4.3)
2. **No stemming** — the paper explicitly states features are "lowercased and unstemmed" (Section 4.3); we follow the same convention
3. **No stopword removal** — stopwords carry signal in deception detection (e.g., pronoun usage); the paper does not remove them
4. **Unit-length normalisation** — document TF-IDF vectors are normalised to unit length (L2 norm), consistent with the paper's SVM setup (Section 4.4: "document vectors are normalized to unit-length")

These steps are applied automatically by `TfidfVectorizer` with `norm='l2'` in Tasks 2.2 and 2.3.

```python
import pandas as pd

# Show the first 5 examples from each class to verify dataset quality
df = pd.DataFrame({'text': texts, 'label': labels, 'class': ['spam' if l==1 else 'ham' for l in labels]})

print("=== Sample SPAM messages (label=1, parallel to deceptive reviews) ===")
for i, row in df[df['label']==1].head(3).iterrows():
    print(f"  [{i}] {row['text'][:80]}")

print()
print("=== Sample HAM messages (label=0, parallel to truthful reviews) ===")
for i, row in df[df['label']==0].head(3).iterrows():
    print(f"  [{i}] {row['text'][:80]}")

print()
print("=== Basic text statistics ===")
df['word_count'] = df['text'].str.split().str.len()
print(df.groupby('class')['word_count'].describe().round(2))
```

**What the code above computes:**  
This cell displays sample messages from each class and computes basic descriptive statistics (word count per class). This mirrors the role of Table 1 in Ott et al. (2011), which reports descriptive statistics for the deceptive opinion submissions. Inspecting word count distributions confirms that both classes have messages of comparable length, ensuring the classifier cannot trivially separate them based on length alone.

**Paper reference:** Section 3, Table 1 — descriptive statistics of deceptive submissions. The authors balance review length distributions between truthful and deceptive classes (Section 3.2); similarly, we verify our two classes have comparable message lengths.
